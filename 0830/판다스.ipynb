```python
import pandas as pd
import numpy as np
```

```python
 # 결측치 np.nan

s1 = pd.Series([1, np.nan, 3 ,4 , 5])
s2 = pd.Series([1, 2 , np.nan , 4 , 5])
s3 = pd.Series([1, 2, 3, np.nan , 5])
```

```python
df = pd.DataFrame({
    'S1' : s1,
    'S2' : s2,
    'S3' : s3
})

df
```

#### NA 값 처리

- dropna( axis  = ) : 결측값 삭제
- isna() : True False 로 반환
- isna().sum() : T/F (1/0)으로 반환 후 sum

```python
df.isna() # 결측치면 True 반환 --> 0 / 1
print(df.isna().sum()) 
```

```python
df.dropna(axis = 0) # 0 = 행 기준으로 na행 삭제 'rows' 도 가능
df.dropna(axis = 1) # 1 = 열 기준으로 na열 삭제 'columns' 도 가능
```

#### 통계 관련 함수

- df.describe() : count, mean, std, minmax, 사분위수

```python
df.describe()
```

```python
##외부의 csv 파일을 로드

sales_df = pd.read_csv('../practice1/data/Sales Records.csv')
```

```python
print(sales_df.describe())
sales_df.isna().sum()
```

```python
#원본데이터 유지
sales_df_copy = sales_df.copy()
```

```python
# 데이터 형태를 datetime으로 변경
print(sales_df_copy['Order Date'])
sales_df_copy['Order Date'] = pd.to_datetime(sales_df_copy['Order Date'], format="%m/%d/%Y")
```

```python
sales_df_copy['Order Date']
```

```python
# Ship Date 컬럼의 데이터를 date time으로 변경

sales_df_copy['Ship Date']
sales_df_copy['Ship Date'] = pd.to_datetime(sales_df_copy['Ship Date'], format="%m/%d/%Y")
print(sales_df_copy['Ship Date'])
```

```python
## 컬럼 이름 변경
sales_df_copy.columns = ['권역','국가', '상품종류', '판매채널', '우선순위', '주문일자', '주문ID', '발송일자', '판매단위', '단가', '원가', '총수익', '총비용', '총이윤']
sales_df_copy.columns
```

```python
# 반올림 기능을 가진 함수 round(n)
# round(n) : n의 값은 소수점 자리수
a = 3.6666
print(round(a,2))
```

```python
sales_df_copy['원가'] = sales_df_copy['원가'].round(0).astype('int')
sales_df_copy.head()
```

```python
## 단가라는 컬럼을 버림으로 해서 int 형으로 변경
## astype('int') 사용

sales_df_copy['단가'] = sales_df_copy['단가'].astype('int')
sales_df_copy.head()
```

```python
##총수익, 총비용, 총이윤 3개의 컬럼 값들을 반올림하여 int형의 데이터로 변경하자

sales_df_copy[['총수익','총비용','총이윤']] = sales_df_copy[['총수익','총비용','총이윤']].round(-1).astype('int')
sales_df_copy.head()
```

#### loc[ ] 함수
: 특정 행/열 추출 시 사용하는 함수
- loc(행의 수, 열의 수/이름 지정)
- 열만 지정하고 싶을 때는 loc[ : , 열의 이름]

```python
sales_df_copy.loc[1:3, "단가":"총수익"]
```

```python
sales_df_copy.loc[:, "단가": "총수익"]
```

#### groupby(컬럼이름).mean().[컬럼 추가 가능]
> 컴럼의 데이터가 같은 값들끼리 그룹화하여 표시 <br>
> mean 대신 묶은 데이터들의 보고 싶은 함수 사용 가능(eg. mean, std, var, min, max etc...)

```python
sales_df_copy.groupby(['국가']).std()[["총이윤","단가"]]
```

```python
sales_df_copy.groupby(["국가"]).get_group('United Kingdom')
```

```python
## 컬럼 및 인덱스 삭제
sales_df_copy.head()
```

#### columns / rows 삭제
> inplace()

```python
##주문 ID 컬럼 삭제
## axis 속성 통해서 인덱스를 삭제할지 컬럼을 삭제할지 지정
## inplace 속성으로 기준이 되는 데이터를 수정할지 지정

sales_df_copy.drop(["주문ID"], axis = 'columns' , inplace=True)
```

#### 파생변수 생성하기

1. 원가, 단가, 판매단위 값을 기준으로 
2. 총수익, 총비용, 총이윤 새로운 컬럼 생성
3. 총수익 = 단가 * 판매단위
4. 총비용 = 원가 * 판매단위
5. 총이윤 = 총수익 - 총비용

##### 새로운 컬럼 추가하기
> dataframe명[새로운 컬럼명] = 컬럼에 들어갈 값(series, list) <br>
> 단가 컬럼 * 판매단위 컬럼


```python
sales_df_copy2["총수익"] = sales_df_copy2["Units Sold"] * sales_df_copy2["Unit Price"] 
sales_df_copy2["총비용"] = sales_df_copy2["Unit Cost"] * sales_df_copy2["Units Sold"]
```

```python
sales_df_copy2["총이윤"] = sales_df_copy2["총수익"] - sales_df_copy2["총비용"]
```
